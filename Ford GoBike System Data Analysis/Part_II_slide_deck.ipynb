## Dataset Overview

> This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area, it contains data for over 183400 bike rides containing information from bike ride duration to type of customer, gender etc for February 2019.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('bike_clean.csv')
```

## Number of Customers to Subscribers

The Ford GoBike company has mainly two types of user types which we all know ,which are the Customers and Subscribers but according to recent times the number of Subscribers has grown considerably Larger than the number of customers with almost a 9:1 ratio.

### Distribution of User Type
There are more subscribers than customers as Subscribers take over 90% of the total dataset while customer takes only 9.5% of the pie chart as shown below

```python
#pie chart showing the percentages of type of customers as a whole
user_counts= df['user_type'].value_counts()
plt.figure(figsize=(8,5))
plt.pie(user_counts, labels = user_counts.index,autopct='%1.1f%%')
plt.title("Pie chart of User types")
plt.axis('square');
```

### Countplot showing number of trips of different user types

The countplot below shows the same information as the pie chart above showing the wide gap between number of subsribers and customers with subscribers count reaching about 160,000 users and customers about 20,000 users

```python
#countplot of number of trips
plt.figure(figsize=(8,5))
default_color = sb.color_palette()[0]
sb.countplot(data=df, x='user_type',color=default_color)
plt.title('Number of Trips by User type',fontsize=13)
plt.xticks(fontsize=11)
plt.xlabel('User type');
```

## User types amongst Gender and Part of day

> It does not matter the gender or the part of the day in which rides are made there are always way more subscribers than customers 

## Number of trips for various user types among gender types
from the bar chart below across both genders we have more subscribers than customers as in the male category we have almost 120,000 subscribers and less than 20,000 customers where as in females we have almost 40,000 subscribers and less than 20,000 customers

```python
#using clustered bar charts to show the relationship between user types and gender
gen_df=df[df['member_gender'].isin(['Male','Female'])]
plt.figure(figsize=(8,5))
sb.countplot(data=gen_df, x='user_type',hue='member_gender')
plt.xticks(fontsize=11)
plt.title('Number of Trips by User type')
plt.xlabel('User type');
```

## Distribution of Number of rides for each part of day using user types

we have in every category large gaps between number of subscribers and customers in every part of day as shown in the bar chart below

```python
plt.figure(figsize=(8,5))
sb.countplot(data = gen_df, x = 'part_of_day', hue = 'user_type')
plt.title('Number of Trips for each part of day using customer type')
plt.xlabel('Part of day');
```

## Average trip Duration
This is the only aspect in which customers get the edge over subscribers as they have a slightly higher average duration of bike rides over subscribers but i belive the number of subscribers to customers balances it out clearly and throws back the advantage to the subscribers  

## Trip Duration for various user types

Subscriber users tend to ride the bikes in lesser duration than customer users with the average mean duration of subscribers being around 500 seconds and customers being around 750 seconds

```python
#using a Boxplot to see the relationships duration ,Gender using User Type
ti=df[df['duration_sec']<=2500]
fig = plt.figure(figsize = [8,6])
sb.boxplot(data = ti, x = 'user_type', y = 'duration_sec', palette = 'autumn')
plt.title('Trip duartion across User Type')
plt.ylabel('Duration (sec)')
plt.xlabel('User Type')
plt.show();
```
